Подключение основных необходимых библиотек

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Загрузка данных о населении, здравоохранении и образовании г. Кургана

```python
df=pd.read_csv("kurgan_data.csv")
```

```python
df.info()
```

Визуализация

Численность населения, прирост, убыль, здравоохранение

```python
datax=df["Year"]
datay=df["Population thousand people"]
plt.plot(datax,datay)
plt.grid()
plt.xlabel("Годы")
plt.ylabel("Численность населения, тыс. чел.")
plt.title("График изменения численности населения по годам")
```

По графику можем выявить пик минимума населения в 2015-2020 годах, пик максимума населения в 1995-2000 годах

```python
datax=df["Year"]
datay=df["Births per thousand people"]
plt.plot(datax,datay)
plt.grid()
plt.xlabel("Годы")
plt.ylabel("Рожденных на тысячу чел.")
plt.title("График изменения прироста населения по годам")
```

По графику можем выявить минимум рождаемости в 1995-2000 годах

```python
datax=df["Year"]
datay=df["Death per thousand people"]
plt.plot(datax,datay)
plt.grid()
plt.xlabel("Годы")
plt.ylabel("Умерших на тысячу чел.")
plt.title("График изменения убыли населения по годам")
```

По графику можем выявить максимум смертности в 2000-2005 годах

Матрица рассеяния по численности населения, рождаемости и смертности

```python
pd.plotting.scatter_matrix(df[["Population thousand people","Births per thousand people","Death per thousand people"]],figsize=(20,15))
```

График зависимостей численности населения, рождаемости и смертности

```python
sns.pairplot(df[["Population thousand people","Births per thousand people","Death per thousand people"]])
```

Графики распределения рождаемости и смертности

```python
sns.distplot(df["Births per thousand people"],rug=True)
```

```python
sns.distplot(df["Death per thousand people"],rug=True)
```

График зависимости рождаемости от количества врачей

```python
sns.relplot(x="Births per thousand people",y="Doctors",data=df)
```

График зависимости рождаемости от количества среднего медицинского персонала

```python
sns.relplot(x="Births per thousand people",y="Nurses",data=df)
```

График зависимости рождаемости от количества медицинских учреждений

```python
sns.relplot(x="Births per thousand people",y="Hospitals",data=df)
```

График зависимостей рождаемости, смертности, количества медицинского персонала и медицинских учреждений

```python
sns.pairplot(df[["Births per thousand people","Death per thousand people","Doctors","Nurses","Hospitals"]])
```

Видна четкая зависимость рождаемости от количества среднего медицинского персонала и менее четкая - от количества медицинских учреждений

Учебные и дошкольные заведения

```python
datax=df["Year"]
datay=df["Pupils"]
plt.plot(datax,datay)
plt.grid()
plt.xlabel("Годы")
plt.ylabel("Детей школьного возраста, тыс. чел.")
plt.title("График изменения числа учащихся по годам")
```

```python
datax=df["Year"]
datay=df["Schools"]
plt.plot(datax,datay)
plt.grid()
plt.xlabel("Годы")
plt.ylabel("Учебные заведения")
plt.title("График изменения числа учебных заведений по годам")
```

Графики распределения детей школьного возраста и школ по годам

```python
sns.distplot(df["Pupils"],rug=True)
```

```python
sns.distplot(df["Schools"],rug=True)
```

График зависимостей численности детей школьного возраста и учебных заведений

```python
sns.pairplot(df[["Pupils","Schools"]])
```

По графикам можем выявить отсутствие взаимосвязи между количеством учащихся и количеством учебных заведений

```python
sns.histplot(x=df["Year"], data=df, hue="Schools", kde=True, bins=6)
```

Видно значительное снижение числа учебных заведений в 2010-2020 года

```python
datax=df["Year"]
datay=df["Preschool childs"]
plt.plot(datax,datay)
plt.grid()
plt.xlabel("Годы")
plt.ylabel("Детей дошкольного возраста, тыс. чел.")
plt.title("График изменения числа детей дошкольного возраста по годам")
```

```python
datax=df["Year"]
datay=df["Preschools"]
plt.plot(datax,datay)
plt.grid()
plt.xlabel("Годы")
plt.ylabel("Дошкольные учреждения")
plt.title("График изменения числа дошкольных учреждений по годам")
```

Графики распределения детей дошкольного возраста и дошкольных учреждений по годам

```python
sns.distplot(df["Preschool childs"],rug=True)
```

```python
sns.distplot(df["Preschools"],rug=True)
```

```python
sns.pairplot(df[["Preschool childs","Preschools"]])
```

Видна зависимость между малым количеством дошкольных заведений и детей дошкольного возраста

```python
sns.histplot(x=df["Year"], data=df, hue="Preschools", kde=True, bins=6)
```

Видно значительное снижение количества дошкольных заведений с 2000 по 2020 годы
